```python
#import library requests
import requests
```

```python
#mengambil informasi dari website api.publicapis.org
req = requests.get('https://api.publicapis.org/entries')
```

```python
#menggunakan public API untuk mengecek ada API publik apa saja yang tersedia berdasarkan kata kunci tertentu
endpoint = "https://api.publicapis.org/entries"
title = "data"

params = {"title": title}
openapi = requests.get(endpoint, params=params).json()
openapi
```

```python
# melakukan pencarian API publik menggunakan kata kunci yang kita inputkan
print('Selamat datang di program pencari API Publik!\n')

endpoint = "https://api.publicapis.org/entries"
query = input("masukkan kata kunci yang ingin anda cari: ")

params = {"title": query}
openapi = requests.get(endpoint, params=params).json()
openapi
```
